```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../preco_medicamentos.csv')
df.head()
```

```python
#Analisando atributos nulos, retornando apenas os que tem dados nulos
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
#Nesse caso, primeiramente irei dropar a coluna 'Análise Recursal'
df = df.drop(columns=['ANÁLISE RECURSAL'])
```

```python
df.isnull().sum()
```

```python
#Agora irei investigar os atributos que possuem - , pois esses são nulos iguais
(df == '-').sum()[ (df == '-').sum() > 0 ]
```

```python
#substituindo o - por nulo, que é oq ele representa
df.replace('-', np.nan, inplace=True)
```

```python
#vendo novamente a quantidade de nulo
df.isnull().sum()
```

```python
#dropando as colunas com mts valores nulos
df.drop(columns=['EAN 2', 'EAN 3'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#Analisando o tipo de dado de cada coluna.
df.dtypes
```

```python
# aqui vou verificar a ocorrencia de cada asterisco nas celulas dos atributos
ocorrencias_asterisco = df.astype(str).apply(lambda x: x.str.contains(r'\*', na=False).sum()) #lista * em qualquer part
ocorrencias_asterisco
```

```python
#  colunas object q contem *
colunas_object = [
    'PF 12%', 'PF 12% ALC', 'PF 17%', 'PF 17% ALC', 'PF 17,5%', 
    'PF 17,5% ALC', 'PF 18%', 'PF 18% ALC', 'PF 19%', 'PF 19% ALC', 
    'PF 19,5%', 'PF 19,5% ALC', 'PF 20%', 'PF 20% ALC', 'PF 20,5%2', 
    'PF 21%', 'PF 21% ALC', 'PF 22%', 'PF 22% ALC',
    'PMVG 12%', 'PMVG 12% ALC', 'PMVG 17%', 'PMVG 17% ALC', 
    'PMVG 17,5%', 'PMVG 17,5% ALC', 'PMVG 18%', 'PMVG 18% ALC', 
    'PMVG 19%', 'PMVG 19% ALC', 'PMVG 19,5%', 'PMVG 19,5% ALC', 
    'PMVG 20%', 'PMVG 20% ALC', 'PMVG 20,5%', 'PMVG 21%', 
    'PMVG 21% ALC', 'PMVG 22%', 'PMVG 22% ALC'
]

# Removendo asteriscos e converter para float
for col in colunas_object:
    if col in df.columns:
        df[col] = df[col].astype(str)  
        df[col] = df[col].str.replace('*', '', regex=False)
        df[col] = df[col].str.replace(',', '.', regex=False)
        df[col] = df[col].astype(float)

# Verificar os tipos de dados após a conversão
print(df.dtypes)
```

```python
# Recalculando a  ocorrencia de cada asterisco 
ocorrencias_asterisco = df.astype(str).apply(lambda x: x.str.contains(r'\*', na=False).sum()) #lista * em qualquer part
ocorrencias_asterisco
```
